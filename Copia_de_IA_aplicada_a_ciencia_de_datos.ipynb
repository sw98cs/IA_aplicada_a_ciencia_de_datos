> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gr√°fico de barras horizontales que muestre la facturaci√≥n por categor√≠a de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categor√≠as de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de env√≠o total de la compra.

Deseamos crear una visualizaci√≥n que muestre la facturaci√≥n de cada categor√≠a, donde la facturaci√≥n se calcula multiplicando el precio unitario por la cantidad y sumando el costo de env√≠o.

Utiliza las bibliotecas de visualizaci√≥n de datos de Python para crear esta visualizaci√≥n.


````python
df
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Crear facturaci√≥n por compra (item + env√≠o)
df = df.copy()
df["facturacion"] = (df["precio_unitario"] * df["cantidad"]) + df["envio"]

# 2) Agrupar por categor√≠a y sumar facturaci√≥n
facturacion_categoria = (
    df.groupby("categoria", as_index=False)["facturacion"]
      .sum()
      .sort_values("facturacion", ascending=True)
)

facturacion_categoria

# 3) Gr√°fico de barras horizontales
ax = plt.gca()
plt.barh(facturacion_categoria["categoria"], facturacion_categoria["facturacion"])
plt.title("Facturaci√≥n por categor√≠a (Zoop)")
plt.xlabel("Facturaci√≥n")
plt.ylabel("Categor√≠a")

# Etiquetas al final de cada barra (opcional)
for i, v in enumerate(facturacion_categoria["facturacion"]):
    plt.text(v, i, f"{v:,.0f}", va="center", ha="left")

plt.tight_layout()
plt.show()
````

### **3a Pregunta:** ¬øC√≥mo fueron las ventas a lo largo de los meses en 2023?


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gr√°fico de l√≠neas que muestre las ventas totales mensuales de Zoop.

En "df", tenemos las siguientes columnas:
* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturaci√≥n: calculada como el producto del precio unitario y la cantidad, sumado al costo de env√≠o de cada venta.

Deseamos crear una visualizaci√≥n que muestre las ventas por mes. Primero, agrupa los datos por mes y luego crea una nueva columna llamada "mes" que contenga el nombre de cada mes, traduciendo los meses del √≠ndice mediante el siguiente diccionario. Utiliza la columna "fecha" para agrupar los datos y construir el gr√°fico. La columna "mes" solo debe usarse para cambiar las etiquetas del eje x.

```
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}
```


Utiliza las bibliotecas de visualizaci√≥n de datos de Python para crear esta visualizaci√≥n.


````python
import pandas as pd
import matplotlib.pyplot as plt

# Diccionario de traducci√≥n
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}

# Asegurar datetime
df["fecha"] = pd.to_datetime(df["fecha"])

# Agrupar por mes usando fecha
ventas_mensuales = (
    df.groupby(pd.Grouper(key="fecha", freq="MS"))["facturacion"]
      .sum()
      .reset_index()
)

# Crear columna "mes" solo para etiquetas del eje x
ventas_mensuales["mes"] = ventas_mensuales["fecha"].dt.strftime("%B").map(meses)

ventas_mensuales

# Gr√°fico de l√≠neas
plt.plot(ventas_mensuales["fecha"], ventas_mensuales["facturacion"], marker="o")
plt.title("Ventas totales mensuales - Zoop")
plt.xlabel("Mes")
plt.ylabel("Facturaci√≥n")

# Cambiar etiquetas del eje x usando la columna "mes"
plt.xticks(ventas_mensuales["fecha"], ventas_mensuales["mes"])

plt.tight_layout()
plt.show()
````

* metodo_pago: con el m√©todo de pago elegido en la compra

````python
def facuracion_por_categoria():

  # 1) Crear facturaci√≥n por compra (item + env√≠o)

  df["facturacion"] = (df["precio_unitario"] * df["cantidad"]) + df["envio"]

  # 2) Agrupar por categor√≠a y sumar facturaci√≥n (ascendente para barh)
  facturacion_categoria = (
      df.groupby("categoria", as_index=False)["facturacion"]
        .sum()
        .sort_values("facturacion", ascending=False)
  )

  # 3) Figura con identidad Zoop
  fig, ax = plt.subplots(figsize=(10, 6))

  # Fondo (figura + √°rea del gr√°fico)
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # Barras horizontales
  bars = ax.barh(
      facturacion_categoria["categoria"],
      facturacion_categoria["facturacion"],
      color=AQUA_1
  )

  # T√≠tulo
  ax.set_title("Facturaci√≥n por categor√≠a de producto en 2023", fontsize=18, color=GRIS_5)

  # Suprimir xticks y labels x/y
  ax.set_xticks([])
  ax.set_xlabel("")
  ax.set_ylabel("")

  # Y ticks (categor√≠as)
  ax.tick_params(axis="y", labelsize=12, colors=AZUL_1)

  # Quitar TODOS los bordes
  for spine in ax.spines.values():
      spine.set_visible(False)

  # Valores a la derecha de cada barra
  max_val = facturacion_categoria["facturacion"].max()
  offset = max_val * 0.01  # separador visual

  for bar in bars:
      w = bar.get_width()
      y = bar.get_y() + bar.get_height() / 2
      ax.text(
          w + offset, y,
          f"MXN${w/1e6:.2f} M",
          va="center",
          ha="left",
          fontsize=12,
          color=AZUL_1
      )

  # 4) Texto dentro del gr√°fico
  # (La 1¬™ barra es la de menor facturaci√≥n porque est√° ordenado ascendente;
  #  aqu√≠ seguimos tu instrucci√≥n literal: 1¬™ barra / suma total.)
  porcentaje = (facturacion_categoria["facturacion"].iloc[0] / facturacion_categoria["facturacion"].sum()) * 100

  ax.text(
      0.4e7, 5,
      f"$\\bf{porcentaje:.2f}$% de las ventas corresponden √∫nicamente a las ventas de Electr√≥nica.\n"
      "Esto equivale a casi el doble en relaci√≥n con todas las dem√°s categor√≠as\n"
      "presentes en el portafolio de Zoop.",
      fontsize=12,
      color=AZUL_3
  )

  plt.tight_layout()
  return plt
````

````python
def ventas_por_trimestre():
  # Asegurar datetime
  df["fecha"] = pd.to_datetime(df["fecha"])

  # 1) Crear columna de trimestre (a√±o-trimestre)
  df["trimestre"] = df["fecha"].dt.to_period("Q").astype(str)  # ej: '2023Q1'

  # 2) Ventas por trimestre y m√©todo de pago (tabla din√°mica)
  ventas_trim_mp = (
      df.pivot_table(
          index="trimestre",
          columns="metodo_pago",
          values="facturacion",
          aggfunc="sum",
          fill_value=0
      )
      .sort_index()
  )

  ventas_trim_mp

  # Formateador eje Y: MXN$ valor M (millones) sin decimales
  def formato_mxn_millones_0(x, pos):
      return f"MXN$ {x/1_000_000:,.0f} M"

  # 3) Figura Zoop
  fig, ax = plt.subplots(figsize=(10, 6))

  # Fondo (figura + √°rea del gr√°fico)
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # Barras apiladas con paleta solicitada
  palette = [BLANCO, GRIS_3, AQUA_1, AQUA_3]
  ventas_trim_mp.plot(kind="bar", stacked=True, ax=ax, color=palette)

  # T√≠tulo
  ax.set_title("Ventas por Trimestre y M√©todo de Pago en 2023", fontsize=18, color=GRIS_5)

  # Suprimir labels de ejes
  ax.set_xlabel("")
  ax.set_ylabel("")

  # X ticks
  ax.tick_params(axis="x", labelrotation=0, labelsize=12, colors=AZUL_1)

  # Y ticks con formato solicitado
  ax.yaxis.set_major_formatter(FuncFormatter(formato_mxn_millones_0))
  ax.tick_params(axis="y", labelsize=12, colors=AZUL_1)

  # Quitar TODOS los bordes
  for side in ["top", "bottom", "left", "right"]:
      ax.spines[side].set_visible(False)

  # Cuadr√≠cula punteada en Y
  ax.grid(axis="y", linestyle=":", linewidth=1)

  # (Opcional) Ajuste de leyenda para que no estorbe
  ax.legend(title="M√©todo de pago", bbox_to_anchor=(1.02, 1), loc="upper left", frameon=False)

  plt.tight_layout()
  return plt
````

````python
def box_plot_edad_sexo():

  # --- Preparaciones (para usar √≠ndices [0] y [1] como pides) ---
  # Ordenar las categor√≠as para que el boxplot quede consistente
  sexos = sorted(df["sexo_biologico"].dropna().unique())
  df_plot = df.copy()
  df_plot["sexo_biologico"] = pd.Categorical(df_plot["sexo_biologico"], categories=sexos, ordered=True)

  # Estad√≠sticos por sexo (usamos el orden de "sexos" para indexar [0] y [1])
  g = df_plot.groupby("sexo_biologico")["edad"]
  mins = g.min()
  q1 = g.quantile(.25)
  med = g.median()
  q3 = g.quantile(.75)
  maxs = g.max()

  # --- Figura Zoop ---
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # Boxplot (controlado con matplotlib)
  bp = ax.boxplot(
      [df_plot.loc[df_plot["sexo_biologico"] == sexos[0], "edad"].dropna(),
      df_plot.loc[df_plot["sexo_biologico"] == sexos[1], "edad"].dropna()],
      patch_artist=True,
      widths=0.5
  )

  # Paleta por sexo: [AZUL_1, AQUA_1]
  palette = [AZUL_1, AQUA_1]
  for patch, c in zip(bp["boxes"], palette):
      patch.set_facecolor(c)
      patch.set_edgecolor(c)

  # Estilo de l√≠neas del boxplot
  for k in ["whiskers", "caps", "medians"]:
      for line in bp[k]:
          line.set_color(GRIS_5)
          line.set_linewidth(1.5)

  for flier in bp["fliers"]:
      flier.set_markeredgecolor(GRIS_5)
      flier.set_markerfacecolor(GRIS_5)
      flier.set_alpha(0.4)

  # T√≠tulo
  ax.set_title(
      "Distribuci√≥n de la Edad por Sexo Biol√≥gico de los  clientes",
      fontsize=18,
      color=GRIS_5
  )

  # Sin xlabel
  ax.set_xlabel("")

  # Ticks y ylabel
  ax.set_xticks([1, 2])
  ax.set_xticklabels(sexos, fontsize=12, color=AZUL_1)
  ax.tick_params(axis="y", labelsize=12, colors=AZUL_1)
  ax.set_ylabel("Edad", fontsize=12, color=AZUL_1)

  # Quitar bordes
  for side in ["top", "bottom", "left", "right"]:
      ax.spines[side].set_visible(False)

  # Grid detr√°s
  ax.set_axisbelow(True)
  ax.grid(axis="y", linestyle=":", linewidth=1)

  # --- Anotaciones ---
  texto_1 = (
      f"M√≠nimo = {mins.iloc[1]}\n"
      f"1¬∫ quartil = {q1.iloc[1]}\n"
      f"2¬∫ quartil = {med.iloc[1]}\n"
      f"3¬∫ quartil = {q3.iloc[1]}\n"
      f"M√°ximo = {maxs.iloc[1]}"
  )

  texto_2 = (
      f"M√≠nimo = {mins.iloc[0]}\n"
      f"1¬∫ quartil = {q1.iloc[0]}\n"
      f"2¬∫ quartil = {med.iloc[0]}\n"
      f"3¬∫ quartil = {q3.iloc[0]}\n"
      f"M√°ximo = {maxs.iloc[0]}"
  )

  ax.annotate(
      texto_1,
      xy=(0.9, 45),
      xytext=(0.6, 50),
      fontsize=12,
      color=AZUL_1,
      bbox=dict(boxstyle="round,pad=0.4", fc=BLANCO, ec=GRIS_3),
      arrowprops=dict(arrowstyle="->", color=GRIS_5, linewidth=1.5)
  )

  ax.annotate(
      texto_2,
      xy=(2, 45),
      xytext=(2.1, 50),
      fontsize=12,
      color=AZUL_1,
      bbox=dict(boxstyle="round,pad=0.4", fc=BLANCO, ec=GRIS_3),
      arrowprops=dict(arrowstyle="->", color=GRIS_5, linewidth=1.5)
  )

  plt.tight_layout()
  return plt
````
